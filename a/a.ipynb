```python
from IPython.core.display import Math
def distance(a, b):
    dist = 0
    for i in range(0, len(a)):
      dist=dist+(a[i]-b[i])**2
    dist=(dist)**(0.5)
    return dist

def normalize(vec):
    normalized=[]
    for i in range(0,len(vec)):
        normal = (vec[i]-min(vec))/(max(vec)-min(vec))
        normalized.append(normal)
    return normalized
```

```python
fruit_name = df["name"].unique()
```

```python
fruit_names = []
fruit_rep = []

for fname in fruit_name:
    fruit_names.append(fname)
    df_f = df.loc[df['name'] == fname]
    list_r = list(df_f["red"])
    list_g = list(df_f["green"])
    list_b = list(df_f.iloc[:,3])
    fg = [np.mean(list_r), np.mean(list_g), np.mean(list_b)]
    fruit_rep.append(fg)

fruits = dict(zip(fruit_names, fruit_rep))
print(fruits)
```

```python
path1 = "Training"
all_Classes = os.listdir(path1) 
prototypes = {}
cl_count = 0
for cl in all_Classes:
    proto = np.zeros((1100))
    count = 0
    red=0
    blue=0
    green=0
    for img in os.listdir(f'{path1}\\{cl}'):
        fruit=imread(f'{path1}\\{cl}\\{img}')
        vector = hog(fruit, orientations=11, pixels_per_cell = (15,15), cells_per_block = (2,2), multichannel=True)
        proto = proto+vector
        count+=1           
    proto = proto/count
    prototypes[cl] = proto
    cl_count+=1
    print("class: ",cl_count ,": ", cl," : done")
```

```python
print(prototypes)
```

```python
path2 = "Test"
classes = os.listdir(path2) 
y_pred =[]
y_true =[]
for cla in classes:
    for img in os.listdir(f'{path2}\\{cla}'):
        fruit=imread(f'{path2}\\{cla}\\{img}')
        
        vector = hog(fruit, orientations=11, pixels_per_cell = (15,15), cells_per_block = (2,2), multichannel=True)
        
        list1 = []
        list2 = []
        for cl in prototypes.keys():
            distance = distance(prototypes[cl],vector)
            list1.append(distance)
            list2.append(cl)
        minim = min(list1)
        minim_ind = list1.index(minim)
        classObjRGB = list2[minim_ind]
        y_true.append(cla)
        y_pred.append(classObjRGB)
        
    print("Class ",cla, "complete" )        
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='micro')
recall = recall_score(y_true, y_pred, average='macro')
F1 = 2*(precision* recall)/(precision + recall)
print("Accuracy = ", accuracy)
print("Precision = ", precision)
print("Recall = ", recall)
print("F1 score = ", F1)
```

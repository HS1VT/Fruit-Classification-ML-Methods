```python
import pandas as pd
import math
data_frame = pd.read_csv('fruits.csv')
```

```python
import cv2
def rgb_test(f):
  r=0
  b=0
  g=0
  count=0
  cap=cv2.imread(f)
  for x in range (0,100,1):
    for y in range(0,100,1):
      color = cap[y,x]
      r=r+color[0]
      g=g+color[1]
      b=b+color[2]
      count=count+1
  r=r/count
  b=b/count
  g=g/count
  l=[r,g,b]
  print(l)
  return l
```

```python
v=rgb_test("wat_test.jpg") # this is the vector of test image of watermelon
test="Watermelon" # We store the value of this so that we can find the accuracy of our model
```

```python
x = data_frame.iloc[:,[1,2,3]].values
print(x)
```

```python
y = data_frame.iloc[:,0].values
print(y)
```

```python
dist=[] #dist contains the euclidean distance of test image from all the other images
for i in range (0,len(y)):
  x1=0
  x2=0
  x3=0
  x1=(x[i][0]-v[0])**2
  x2=(x[i][1]-v[1])**2
  x3=(x[i][2]-v[2])**2
  d=math.sqrt(x1+x2+x3)
  dist.append(d)
```

```python
print(dist)
```

```python
N=11 # we find the indices of 11 minumum elements from the list
res = sorted(range(len(dist)), key = lambda sub: dist[sub], reverse=True)[-N:]
print(res)
```

```python
y_dist=[] # this list contains the top 11 fruits that our model predicted the test image to be
for j in range(0,len(res)):
  y_dist.append(y[res[j]])
print(y_dist)
```

```python
from pandas.core.arrays.numeric import T
yes=0
no=0
for k in range(0, len(y_dist)):
  if y_dist[k]==test:
    yes=yes+1
  else:
    no=no+1
accuarcy=((yes)/(yes+no))*100
print("The accuracy of the model is",accuarcy,"% for",test)
```

```python
# We see that our model is 100% accurate in predicting watermelon as the top 10 matches are all watermelon
# We can repeat the same for any fruit we want
# We can conclude that our code works pretty well for Watermelon :)
```
